```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = df.iloc[:,[1,2,6,7,8,9,10,11]]
data=pd.get_dummies(data)
```

```python
x,y=data.iloc[:,1:16], data.iloc[:,0]
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42)
```

```python
lnr=LinearRegression().fit(x_train, y_train)
```

```python
print("lnr.coef_: {}".format(lnr.coef_))
print("lnr.intercept_: {}".format(lnr.intercept_))
```

```python
print("Training Score: {:.2f}".format(lnr.score(x_train, y_train)))
print("Test Score: {:.2f}".format(lnr.score(x_train, y_train)))
```
